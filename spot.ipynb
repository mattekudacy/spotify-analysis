<h1> Audio Features </h1>
<ol type="1">
    <li style="padding-bottom:10px"><b>Acousticness</b> - a confidence measure that a track is acoustic. 1.0 being the highest confidence</li>
    <li style="padding-bottom:10px"><b>Danceability</b> - describes how suitable a song is for dancing based on a combination of musical elements including tempo, rhythm, stability, beat strength, and overall regularity. 1.0 confidence is the most danceable. </li>
    <li style="padding-bottom:10px"><b>Duration(ms)</b> - duration of track</li>
    <li style="padding-bottom:10px"><b>Energy</b> - a measure of 0.0 to 1.0 that represents a perceptual measure of intesity and activity.</li>
    <li style="padding-bottom:10px"><b>Instrumentals</b> - assuems that a track has no vocals.</li>
    <li style="padding-bottom:10px"><b>Key</b> - Integers map to pitches using standard Pitch Class notation . E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on.</li>
    <li style="padding-bottom:10px"><b>Liveness</b> - presence of a live audience in the track</li>
    <li style="padding-bottom:10px"><b>Loudness</b> - Loudness of the track in decibels(db)</li>
    <li style="padding-bottom:10px"><b>Mode</b> -  indicates the modality (major or minor) of a track.</li>
    <li style="padding-bottom:10px"><b>Speechiness</b> - assuems that a track has no vocals.</li>
    <li style="padding-bottom:10px"><b>Tempo</b> - The overall estimated tempo of a track in beats per minute (BPM)</li>
    <li style="padding-bottom:10px"><b>Time Signature</b> - assuems that a track has no vocals.</li>
    <li style="padding-bottom:10px"><b>Valence</b> -  A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track.</li>
    <li style="padding-bottom:10px"><b>Liked</b> -  1 for liked, 0 for disliked.</li>
</ol>

<h1> Import Libraries

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

<h1> Display Cleaned Data

```python
with open("like.json", "r") as f:
    liked = json.load(f)
liked = pd.DataFrame(liked['audio_features'])
liked
```

```python
with open("dislike.json", "r") as f:
    disliked = json.load(f)
disliked = pd.DataFrame(disliked['audio_features'])
disliked
```

<h1> Add Liked and Disliked Columns

```python
liked['liked'] = [1] * 100
disliked['liked'] = [0] * 100
liked
```

```python
disliked
```

```python
data = pd.concat([liked, disliked])
data
```

<h1> Remove Irrevelevant columns

```python
data.drop(["type","id","uri","track_href","analysis_url"],axis=1,inplace=True)
data
```

<h1> Shuffle data

```python
data = data.sample(frac=1)
data
```
